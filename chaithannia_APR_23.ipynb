```python
q1=td['Fare'].quantile(0.25)
q3=td['Fare'].quantile(0.75)
iqr=q3-q1
iqr
```

```python
upper_lim=q3+1.5*iqr
low_lim=q1-1.5*iqr
print(low_lim)
print(upper_lim)
```

```python
out=[]
for x in td['Fare']:
  if x>upper_lim or x < low_lim:
    out.append(x)
out

```

```python
td['Fare'] = td['Fare'].clip(lower=low_lim,upper=upper_lim)
```
