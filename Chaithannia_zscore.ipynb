```python
import numpy as np
import pandas as pd
td=pd.read_csv('/content/beer-servings.csv')
td.head()
```

```python
td.columns
```

```python
td2=td.copy()
```

```python
td2['beer_servings'].fillna(value=td2['beer_servings'].mean(),inplace=True)
td2['spirit_servings'].fillna(value=td2['spirit_servings'].mean(),inplace=True)
td2['wine_servings'].fillna(value=td2['wine_servings'].mean(),inplace=True)
td2['total_litres_of_pure_alcohol'].fillna(value=td2['total_litres_of_pure_alcohol'].mean(),inplace=True)
```

```python
td2.isna().sum()
```

```python
import scipy.stats as stats
td2['z_scores1'] = stats.zscore(td2['beer_servings'])
td2['z_scores2'] = stats.zscore(td2['spirit_servings'])
td2['z_scores3'] = stats.zscore(td2['wine_servings'])
td2['z_scores4'] = stats.zscore(td2['total_litres_of_pure_alcohol'])
td2.head()

```

```python
z_scores1=(np.abs(td2['z_scores1'])>3)

z_scores1.unique()
```

```python
z_scores2=(np.abs(td2['z_scores2'])>3)

z_scores2.value_counts()
```

```python
z_scores3=(np.abs(td2['z_scores3'])>3)

z_scores3.value_counts()
```

```python
z_scores4=(np.abs(td2['z_scores4'])>3)

z_scores4.value_counts()
```

**Shows outliers with in 2 columns spirit_servings and wine_servings**
