```python
#scaling

from sklearn.preprocessing import StandardScaler
Std_ad = StandardScaler()
for i in ad1.columns:
    ad1[i] = Std_ad.fit_transform(ad1[i].values.reshape(-1,1))
Std_ad.fit_transform(ad1)
ad1.head(5)
```

**Kmeans clustering**

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=42)
  kmeans.fit(ad1)
  wcss.append(kmeans.inertia_)
wcss
```

```python
plt.figure(figsize=(6,6))
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker='o')
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.title("Elbow method")
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
from sklearn.metrics import silhouette_score
kmeans_silh_score=[]
for k in range(2,11):
  kmeans=KMeans(n_clusters=k,max_iter=300,n_init=10,random_state=42)
  kmeans_labels=kmeans.fit_predict(ad1)
  silh_avg=silhouette_score(ad1,kmeans_labels)
  kmeans_silh_score.append(silh_avg)
```

```python
for k,score in enumerate(kmeans_silh_score,start=2):
  print(f"number of clusters = {k} and silhouette score = {score}")
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(2,11),kmeans_silh_score,marker='o')
plt.title("kmeans_silh_score")
plt.xlabel("no of clusters")
plt.ylabel("silh_score")
plt.show()
```

```python
#take the point where it meets the optimum point ,here we can take it as 3
kmeans=KMeans(n_clusters=3,init="k-means++",max_iter=300,n_init=10,random_state=42)
kmeans.fit(ad1)
cluster_labels=kmeans.predict(ad1)
cluster_labels
```

```python
ad2=ad1.copy()
ad2['Cluster'] =cluster_labels
plt.figure(figsize=(6,6))
sns.scatterplot(x='Alcohol',y='Malic_Acid',data=ad2,hue='Cluster', palette =['blue','green','red'], s=50)
plt.xlabel("'Alcohol'")
plt.ylabel("'Malic_Acid' ")
plt.title("'Alcohol' vs Malic_Acid")
plt.show()
```

**HC method-Agglomarative clustering**

```python
from scipy.cluster.hierarchy import linkage,dendrogram
mergings=linkage(ad1,method='ward')
dendrogram(mergings)
plt.title("Dendrogram")
plt.xlabel("customers")
plt.show()
```

```python
# n_cluster will be 3
ad3=ad1.copy()
from sklearn.cluster import AgglomerativeClustering
agg_hc=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
y_agg_hc=agg_hc.fit_predict(ad3)
```

# **Principal Component analysis**-PCA


```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
pca.fit(ad)
ad_pca=pca.transform(ad)
ad_pca.shape
```

**HC-after PCA**

```python
from scipy.cluster.hierarchy import linkage,dendrogram
mergings=linkage(ad_pca,method='ward')
dendrogram(mergings)
plt.title("Dendrogram")
plt.xlabel("customers")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_hc=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
y_agg_hc=agg_hc.fit_predict(ad_pca)
```

```python
ad3['Cluster_pca']=y_agg_hc
ad3['Cluster_pca'].value_counts()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='Alcohol',y='Malic_Acid',data=ad3,hue='Cluster_pca', palette =['blue','green','red'], s=50)
plt.xlabel("'Alcohol'")
plt.ylabel("'Malic_Acid' ")
plt.title("'Alcohol' vs Malic_Acid")
plt.show()
```

**Kmeans after PCA**

```python
ad5=ad_pca.copy()
kmeans=KMeans(n_clusters=3,init="k-means++",max_iter=300,n_init=10,random_state=42)
kmeans.fit(ad5)
cluster_labels=kmeans.predict(ad5)
cluster_labels
```

```python
ad3['Cluster1'] =cluster_labels
plt.figure(figsize=(6,6))
sns.scatterplot(x='Alcohol',y='Malic_Acid', data=ad3,
                hue='Cluster1', palette =['blue','green','red'], s=50)
plt.xlabel("Alcohol")
plt.ylabel("Malic_Acid")
plt.title("Alcohol vs Malic_Acid")
plt.show()
```

# **DBSCAN algorithm implementation-trial**

```python
x=ad1[['Alcohol','Malic_Acid']]
from sklearn.neighbors import NearestNeighbors
nb=NearestNeighbors(n_neighbors=4)
nb_fit=nb.fit(x)
dist,indices=nb_fit.kneighbors(x)
dist=np.sort(dist,axis=1)
dist=dist[:,3]
plt.plot(dist)

```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.8,min_samples=10).fit(x)
ad3['db_clusters']=dbscan.labels_
ad3
```
